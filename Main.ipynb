```
def clean(text):
    text = str(text).lower()
    rmv = {}
    for i in range(len(text)):
        if text[i]=='<':
            tmp = ''
            while text[i]!='>': 
                tmp += text[i]
                i += 1
            rmv[tmp+'>']=1
    for w in list(rmv.keys()):
        text = text.replace(w, '')
    return annotator.tokenize(text)
```

```
class Count:
    def __init__(self):
        self.culture = 0
        self.fate = 0
        self.sex = 0
        self.violent = 0
```

```
docs = pd.read_csv("output.csv")
```

```
base_tags = {'culture':0, 'cultist':0, 'sex':0, 'violent':0}
```

```
# docs = data[:1000]
domain = {}
drops = []
for i in range(docs.shape[0]):
    try:
        print(f'Executing {i+1}/{docs.shape[0]}')
        docs['message'][i] = clean(docs['message'][i])
        domain[docs['domain'][i]] = {'culture':0, 'cultist':0, 'sex':0, 'violent':0}
    except:
        drops.append(i)
docs.drop(drops)
```

```
vocab = {}
for i in range(docs.shape[0]):
    for sent in docs['message'][i]:
        for word in sent:
            if word not in vocab: vocab[word] = 1
len(vocab)
```

```
import re

for word in list(vocab.keys()):
    if re.findall(r"['\[\]]", word) or re.findall(r"[0-9]|[\\$&+,:;=?@#|<>.\^*()%!/\-]", repr(word)): vocab.pop(word)
len(vocab)
```

```
stop_words = []
with open('stopword.txt') as f:
    stop_words = (f.read().split())
for word in stop_words:
    if word in vocab: vocab.pop(word)
len(vocab)
```

```
vocab = sorted(vocab.keys())
print(vocab)
```

```
key_words = {'cultist':{}, 'culture':{}, 'sex':{}, 'violent':{}}
for tag in list(base_tags.keys()):
    with open(f'{tag}.txt') as f: key_words[tag] = f.read().split()
key_words
```

```
table = []
for i in range(docs.shape[0]):
    print(f'Counting doc {i+1}/{docs.shape[0]}')
    doc = dict.fromkeys(vocab, 0)
    size = 0
    for sent in docs['message'][i]:
        for word in sent: 
            if word in doc: 
                doc[word] += 1
                size += 1
    doc['size_of_doc'] = size
    table.append(doc)
df = pd.DataFrame(table)
df
```

```
import math

size = df['size_of_doc']*float(1)
idf = df[df>0].count().apply(lambda x: math.log10(docs.shape[0]/(float(x)+1)))
# for i in range(docs.shape[0]):
#     print(f'Tf-idf calcutating doc {i+1}/{docs.shape[0]}')
#     for word in df.keys():
#         if word!='size_of_doc': df.at[i, word] = df.at[i, word]/float(df.at[i, 'size_of_doc'])*idf[word]
df = df.div(size, axis='index')
df = df.mul(idf, axis='columns')
df
```

```
df = df.T
df
```

```
tags = ['culture', 'cultist', 'sex', 'violent']
test = []
for i in range(docs.shape[0]):
    series = df[i]
    top = series.nlargest(math.floor(0.05*size[i]), keep='all')
    # print(top)
    for word in top.keys():
        for tag in tags:
            if word in key_words[tag]:
                test.append([tag, docs['message'][i]])
                domain[docs['domain'][i]][tag] += 1
                print(domain)
                break
test = pd.DataFrame(test)
test.to_excel('Checking.xlsx')
```
